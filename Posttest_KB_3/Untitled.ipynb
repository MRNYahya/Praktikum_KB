```python
df = pd.read_csv("anime.csv")

fig = plt.figure(figsize = (15,8))
ax = fig.add_axes([0,0,1,1])

plt.title("Grafik Rating Anime")
ax.set_xlabel("Judul Anime")
ax.set_ylabel("Rating")
ax.plot(df.name[0:5],df.rating[0:5], color='red')
plt.grid(True)
```

Keterangan: dapat dilihat dari grafik di atas, anime mana yang memiliki rating lebih bagus daripada anime lainnya

```python
df.to_csv("rating.csv")
```

# PIE CHART

```python
data_pie = df['type'].value_counts().rename_axis('type').reset_index(name='Id_count')
plt.figure(figsize=(30,26))
colors = sns.color_palette('Set2', 8)
plt.pie(data_pie.Id_count, labels=data_pie.type, colors=colors, startangle=90, autopct='%.1f%%')
plt.title('Type')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

Keterangan: Berdasarkan grafik pie chart di atas, kita dapat melihat type anime apa yang paling banyak dibuat. Dan menurut hasil grafiknya menunjukkan anime yang ber type TV lebih banyak dibuat ketimbang type type lainnya

```python
df.to_csv("type.csv")
```
